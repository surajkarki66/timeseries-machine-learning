## 1. Finding Trend from Timeseries

```python
from sktime import datasets
from sktime.utils.plotting import plot_series
```

```python
airlines = datasets.load_airline()
```

```python
airlines.head()
```

```python
plot_series(airlines)
```

```python
import pandas as pd
```

```python
data = pd.DataFrame({"number": airlines.values})
```

```python
data.head()
```

```python
moving_avg = data.number.rolling(10).mean()
```

```python
plot_series(moving_avg)
```

```python
moving_avg = data.number.rolling(20).mean()
```

```python
plot_series(moving_avg)
```

```python
moving_avg = data.number.rolling(50).mean()
```

```python
plot_series(moving_avg)
```
